```python
from sklearn.model_selection import train_test_split

# Handle missing values
print(f"Missing ABV values: {df['beer_abv'].isnull().sum()}")
df = df.dropna(subset=['beer_abv'])  # Drop rows with missing ABV
print(f"Dataset size after handling missing values: {df.shape}")

# Check for duplicates
duplicates = df.duplicated(subset=['review_profilename', 'beer_beerid'])
print(f"Number of duplicate reviews: {duplicates.sum()}")
df = df.drop_duplicates(subset=['review_profilename', 'beer_beerid'])
print(f"Dataset size after removing duplicates: {df.shape}")

# Filter for users and beers with minimum number of reviews
# This helps address the cold start problem
min_reviews = 3  # Minimum number of reviews per user/beer

# Count reviews per user and beer
user_reviews = df['review_profilename'].value_counts()
beer_reviews = df['beer_beerid'].value_counts()

# Filter users and beers with sufficient reviews
users_to_keep = user_reviews[user_reviews >= min_reviews].index
beers_to_keep = beer_reviews[beer_reviews >= min_reviews].index

# Apply filters
filtered_df = df[df['review_profilename'].isin(users_to_keep) & df['beer_beerid'].isin(beers_to_keep)]
print(f"Dataset size after filtering: {filtered_df.shape}")
print(f"Number of unique users: {filtered_df['review_profilename'].nunique()}")
print(f"Number of unique beers: {filtered_df['beer_beerid'].nunique()}")

# Create user and beer indices for matrix operations
user_ids = filtered_df['review_profilename'].unique()
beer_ids = filtered_df['beer_beerid'].unique()

user_to_idx = {user: i for i, user in enumerate(user_ids)}
beer_to_idx = {beer: i for i, beer in enumerate(beer_ids)}

filtered_df['user_idx'] = filtered_df['review_profilename'].map(user_to_idx)
filtered_df['beer_idx'] = filtered_df['beer_beerid'].map(beer_to_idx)

# Split data into train and test sets (80/20 split)
train_df, test_df = train_test_split(filtered_df, test_size=0.2, random_state=42)
print(f"Training set size: {train_df.shape}")
print(f"Test set size: {test_df.shape}")

# Create rating matrices
n_users = len(user_ids)
n_beers = len(beer_ids)

# Initialize empty matrices
train_matrix = np.zeros((n_users, n_beers))
test_matrix = np.zeros((n_users, n_beers))

# Fill training matrix
for _, row in train_df.iterrows():
    train_matrix[row['user_idx'], row['beer_idx']] = row['review_overall']

# Fill test matrix
for _, row in test_df.iterrows():
    test_matrix[row['user_idx'], row['beer_idx']] = row['review_overall']

print(f"Training matrix shape: {train_matrix.shape}")
print(f"Test matrix shape: {test_matrix.shape}")
print(f"Sparsity: {1 - (np.count_nonzero(train_matrix) / (n_users * n_beers)):.4f}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def user_based_cf(train_matrix, user_idx, n_neighbors=10, top_n=5):
    """
    User-based collaborative filtering

    Parameters:
    -----------
    train_matrix : numpy.ndarray
        The user-item rating matrix
    user_idx : int
        The index of the target user
    n_neighbors : int
        Number of similar users to consider
    top_n : int
        Number of recommendations to return

    Returns:
    --------
    list
        Top N recommended item indices
    """
    # Compute user similarity matrix
    user_similarity = cosine_similarity(train_matrix)

    # Get similarity scores for the target user
    similarities = user_similarity[user_idx]

    # Get indices of similar users (excluding the target user)
    similar_users = np.argsort(similarities)[::-1][1:n_neighbors+1]

    # Get the ratings of similar users
    similar_user_ratings = train_matrix[similar_users]

    # Calculate weighted ratings based on similarity
    user_weights = similarities[similar_users].reshape(-1, 1)
    weighted_ratings = similar_user_ratings * user_weights

    # Sum weighted ratings and normalize
    predicted_ratings = weighted_ratings.sum(axis=0) / (user_weights.sum() + 1e-10)

    # Get items the user hasn't rated yet
    user_ratings = train_matrix[user_idx]
    unrated_items = np.where(user_ratings == 0)[0]

    # Get top N items with highest predicted ratings
    item_scores = [(item, predicted_ratings[item]) for item in unrated_items]
    item_scores.sort(key=lambda x: x[1], reverse=True)

    top_items = [item for item, _ in item_scores[:top_n]]

    return top_items

# Test the user-based CF function
test_user_idx = 10  # Example user index
recommendations = user_based_cf(train_matrix, test_user_idx)
print(f"Top 5 beer recommendations for user {test_user_idx}:")
for i, beer_idx in enumerate(recommendations):
    beer_id = beer_ids[beer_idx]
    beer_name = filtered_df[filtered_df['beer_beerid'] == beer_id]['beer_name'].iloc[0]
    print(f"{i+1}. {beer_name}")
```

```python
def item_based_cf(train_matrix, user_idx, n_neighbors=10, top_n=5):
    """
    Item-based collaborative filtering

    Parameters:
    -----------
    train_matrix : numpy.ndarray
        The user-item rating matrix
    user_idx : int
        The index of the target user
    n_neighbors : int
        Number of similar items to consider
    top_n : int
        Number of recommendations to return

    Returns:
    --------
    list
        Top N recommended item indices
    """
    # Compute item similarity matrix
    item_similarity = cosine_similarity(train_matrix.T)

    # Get the user's ratings
    user_ratings = train_matrix[user_idx]

    # Items the user has rated
    rated_items = np.where(user_ratings > 0)[0]

    # Items the user hasn't rated
    unrated_items = np.where(user_ratings == 0)[0]

    # Calculate predicted ratings
    predicted_ratings = np.zeros(len(unrated_items))

    for i, item_idx in enumerate(unrated_items):
        # Get similarity scores between this item and all items the user has rated
        item_similarities = item_similarity[item_idx, rated_items]

        # Get the user's ratings for similar items
        similar_item_ratings = user_ratings[rated_items]

        # Calculate weighted average rating
        if item_similarities.sum() > 0:
            predicted_ratings[i] = np.sum(item_similarities * similar_item_ratings) / np.sum(item_similarities)
        else:
            predicted_ratings[i] = 0

    # Get top N items with highest predicted ratings
    top_indices = np.argsort(predicted_ratings)[::-1][:top_n]
    top_items = [unrated_items[idx] for idx in top_indices]

    return top_items

# Test the item-based CF function
test_user_idx = 10  # Example user index
recommendations = item_based_cf(train_matrix, test_user_idx)
print(f"\nTop 5 beer recommendations for user {test_user_idx} (item-based CF):")
for i, beer_idx in enumerate(recommendations):
    beer_id = beer_ids[beer_idx]
    beer_name = filtered_df[filtered_df['beer_beerid'] == beer_id]['beer_name'].iloc[0]
    print(f"{i+1}. {beer_name}")
```

```python
# Implement Matrix Factorization using Non-negative Matrix Factorization (NMF)
def matrix_factorization(train_matrix, user_idx, n_components=20, top_n=5):
    """
    Matrix factorization recommendation using NMF

    Parameters:
    -----------
    train_matrix : numpy.ndarray
        The user-item rating matrix
    user_idx : int
        The index of the target user
    n_components : int
        Number of latent factors
    top_n : int
        Number of recommendations to return

    Returns:
    --------
    list
        Top N recommended item indices
    """
    # Replace zeros with NaN to ignore them during factorization
    train_matrix_nan = train_matrix.copy()
    train_matrix_nan[train_matrix_nan == 0] = np.nan

    # Initialize and fit NMF model
    model = NMF(n_components=n_components, init='random', random_state=42, max_iter=200)

    # Fill NaN with mean rating to make NMF work
    mean_rating = np.nanmean(train_matrix_nan)
    train_matrix_filled = np.nan_to_num(train_matrix_nan, nan=mean_rating)

    # Fit the model
    user_factors = model.fit_transform(train_matrix_filled)
    item_factors = model.components_

    # Get user's rated items
    user_ratings = train_matrix[user_idx]
    unrated_items = np.where(user_ratings == 0)[0]

    # Predict ratings for unrated items
    user_embedding = user_factors[user_idx].reshape(1, -1)
    predicted_ratings = np.dot(user_embedding, item_factors)

    # Create a list of (item_idx, predicted_rating) tuples for unrated items
    item_ratings = [(item, predicted_ratings[0, item]) for item in unrated_items]

    # Sort by predicted rating and get top N
    item_ratings.sort(key=lambda x: x[1], reverse=True)
    top_items = [item for item, _ in item_ratings[:top_n]]

    return top_items

# Test the matrix factorization function
test_user_idx = 10  # Example user index
recommendations = matrix_factorization(train_matrix, test_user_idx)
print(f"\nTop 5 beer recommendations for user {test_user_idx} (matrix factorization):")
for i, beer_idx in enumerate(recommendations):
    beer_id = beer_ids[beer_idx]
    beer_name = filtered_df[filtered_df['beer_beerid'] == beer_id]['beer_name'].iloc[0]
    print(f"{i+1}. {beer_name}")
```

```python
def content_based_recommender(filtered_df, user_idx, top_n=5):
    """
    Content-based beer recommender

    Parameters:
    -----------
    filtered_df : pandas.DataFrame
        The beer reviews dataframe
    user_idx : int
        The index of the target user
    top_n : int
        Number of recommendations to return

    Returns:
    --------
    list
        Top N recommended beer indices
    """
    # Get user profile name
    user_name = user_ids[user_idx]

    # Get beers rated by this user
    user_ratings = filtered_df[filtered_df['review_profilename'] == user_name]

    # If user has no ratings, return popular beers
    if len(user_ratings) == 0:
        top_beers = filtered_df.groupby('beer_idx')['review_overall'].mean().nlargest(top_n).index.tolist()
        return top_beers

    # Get positively rated beers (above average for this user)
    user_avg_rating = user_ratings['review_overall'].mean()
    liked_beers = user_ratings[user_ratings['review_overall'] >= user_avg_rating]['beer_beerid'].tolist()

    # If no liked beers, return popular beers
    if len(liked_beers) == 0:
        top_beers = filtered_df.groupby('beer_idx')['review_overall'].mean().nlargest(top_n).index.tolist()
        return top_beers

    # Get beer features for content-based filtering
    beer_features = filtered_df[['beer_beerid', 'beer_style', 'brewery_id', 'beer_abv']].drop_duplicates('beer_beerid')

    # One-hot encode categorical features
    style_encoder = OneHotEncoder(sparse_output=False)
    brewery_encoder = OneHotEncoder(sparse_output=False)

    style_encoded = style_encoder.fit_transform(beer_features[['beer_style']])
    brewery_encoded = brewery_encoder.fit_transform(beer_features[['brewery_id']])

    # Scale ABV
    scaler = StandardScaler()
    abv_scaled = scaler.fit_transform(beer_features[['beer_abv']])

    # Combine features
    features = np.hstack([style_encoded, brewery_encoded, abv_scaled])

    # Calculate similarity between beers
    beer_similarity = cosine_similarity(features)

    # Get indices of beers the user has liked
    beer_indices = beer_features['beer_beerid'].tolist()
    liked_indices = [beer_indices.index(beer_id) for beer_id in liked_beers]

    # Get all beer indices
    all_indices = range(len(beer_indices))

    # Calculate similarity scores for all beers
    similarity_scores = np.zeros(len(beer_indices))

    for idx in liked_indices:
        similarity_scores += beer_similarity[idx]

    # Create (beer_idx, similarity_score) pairs
    beer_scores = [(i, similarity_scores[i]) for i in all_indices
                   if beer_indices[i] not in liked_beers]  # Exclude already rated beers

    # Sort by similarity score
    beer_scores.sort(key=lambda x: x[1], reverse=True)

    # Get top N beers
    top_beer_indices = [beer_indices[idx] for idx, _ in beer_scores[:top_n]]

    # Convert beer IDs to beer indices in the matrix
    top_beer_matrix_indices = [beer_to_idx[beer_id] for beer_id in top_beer_indices]

    return top_beer_matrix_indices

# Test the content-based recommender
test_user_idx = 10  # Example user index
recommendations = content_based_recommender(filtered_df, test_user_idx)
print(f"\nTop 5 beer recommendations for user {test_user_idx} (content-based):")
for i, beer_idx in enumerate(recommendations):
    beer_id = beer_ids[beer_idx]
    beer_info = filtered_df[filtered_df['beer_beerid'] == beer_id].iloc[0]
    print(f"{i+1}. {beer_info['beer_name']} ({beer_info['beer_style']}, ABV: {beer_info['beer_abv']}%)")
```

```python
def hybrid_recommender(train_matrix, filtered_df, user_idx, top_n=5, weights=[0.4, 0.3, 0.3]):
    """
    Hybrid recommendation system combining collaborative filtering and content-based approaches

    Parameters:
    -----------
    train_matrix : numpy.ndarray
        The user-item rating matrix
    filtered_df : pandas.DataFrame
        The beer reviews dataframe
    user_idx : int
        The index of the target user
    top_n : int
        Number of recommendations to return
    weights : list
        Weights for user-based CF, item-based CF, and content-based recommendations

    Returns:
    --------
    list
        Top N recommended beer indices
    """
    # Check if user has ratings
    user_ratings = train_matrix[user_idx]
    has_ratings = np.sum(user_ratings > 0) > 0

    # For cold start problem (new users)
    if not has_ratings:
        # Return popular beers based on average ratings
        beer_avg_ratings = filtered_df.groupby('beer_idx')['review_overall'].mean()
        top_beers = beer_avg_ratings.nlargest(top_n).index.tolist()
        return top_beers

    # Get recommendations from each model
    try:
        user_cf_recs = user_based_cf(train_matrix, user_idx, top_n=top_n*2)
    except:
        user_cf_recs = []

    try:
        item_cf_recs = item_based_cf(train_matrix, user_idx, top_n=top_n*2)
    except:
        item_cf_recs = []

    try:
        content_recs = content_based_recommender(filtered_df, user_idx, top_n=top_n*2)
    except:
        content_recs = []

    # Combine recommendations with weights
    all_recs = {}

    # Add user-based CF recommendations with weight
    for i, beer_idx in enumerate(user_cf_recs):
        if beer_idx not in all_recs:
            all_recs[beer_idx] = 0
        all_recs[beer_idx] += weights[0] * (1 - i/(len(user_cf_recs) + 1))

    # Add item-based CF recommendations with weight
    for i, beer_idx in enumerate(item_cf_recs):
        if beer_idx not in all_recs:
            all_recs[beer_idx] = 0
        all_recs[beer_idx] += weights[1] * (1 - i/(len(item_cf_recs) + 1))

    # Add content-based recommendations with weight
    for i, beer_idx in enumerate(content_recs):
        if beer_idx not in all_recs:
            all_recs[beer_idx] = 0
        all_recs[beer_idx] += weights[2] * (1 - i/(len(content_recs) + 1))

    # Sort by weighted score
    sorted_recs = sorted(all_recs.items(), key=lambda x: x[1], reverse=True)

    # Get top N recommendations
    top_recs = [beer_idx for beer_idx, _ in sorted_recs[:top_n]]

    return top_recs

# Test the hybrid recommender
test_user_idx = 10  # Example user index
recommendations = hybrid_recommender(train_matrix, filtered_df, test_user_idx)
print(f"\nTop 5 beer recommendations for user {test_user_idx} (hybrid):")
for i, beer_idx in enumerate(recommendations):
    beer_id = beer_ids[beer_idx]
    beer_info = filtered_df[filtered_df['beer_beerid'] == beer_id].iloc[0]
    print(f"{i+1}. {beer_info['beer_name']} ({beer_info['beer_style']}, ABV: {beer_info['beer_abv']}%)")
```

```python
def predict_rating(train_matrix, test_matrix, method='user_cf', n_neighbors=10, n_components=20):
    """
    Predict ratings using different methods and calculate error metrics

    Parameters:
    -----------
    train_matrix : numpy.ndarray
        The training user-item rating matrix
    test_matrix : numpy.ndarray
        The test user-item rating matrix
    method : str
        Method to use ('user_cf', 'item_cf', 'matrix_fact', 'baseline')
    n_neighbors : int
        Number of neighbors for collaborative filtering
    n_components : int
        Number of components for matrix factorization

    Returns:
    --------
    tuple
        (RMSE, MAE, Coverage)
    """
    # Get non-zero indices in test matrix (actual ratings)
    test_indices = np.where(test_matrix > 0)
    n_test = len(test_indices[0])

    # Initialize arrays for predictions
    predictions = np.zeros(n_test)
    actual = np.zeros(n_test)
    covered = np.zeros(n_test, dtype=bool)

    # Baseline: global mean
    if method == 'baseline':
        global_mean = np.mean(train_matrix[train_matrix > 0])
        predictions.fill(global_mean)
        actual = test_matrix[test_indices]
        covered.fill(True)

    # User-based CF
    elif method == 'user_cf':
        for i in range(n_test):
            user_idx = test_indices[0][i]
            item_idx = test_indices[1][i]
            actual[i] = test_matrix[user_idx, item_idx]

            # Get user similarity
            user_similarity = cosine_similarity(train_matrix)[user_idx]
            similar_users = np.argsort(user_similarity)[::-1][1:n_neighbors+1]

            # Get ratings from similar users
            sim_user_ratings = train_matrix[similar_users, item_idx]
            sim_weights = user_similarity[similar_users]

            # If there are ratings for this item from similar users
            if np.sum(sim_user_ratings > 0) > 0:
                mask = sim_user_ratings > 0
                if np.sum(mask) > 0:
                    weighted_sum = np.sum(sim_user_ratings[mask] * sim_weights[mask])
                    weight_sum = np.sum(sim_weights[mask])
                    if weight_sum > 0:
                        predictions[i] = weighted_sum / weight_sum
                        covered[i] = True

    # Item-based CF
    elif method == 'item_cf':
        for i in range(n_test):
            user_idx = test_indices[0][i]
            item_idx = test_indices[1][i]
            actual[i] = test_matrix[user_idx, item_idx]

            # Get item similarity
            item_similarity = cosine_similarity(train_matrix.T)[item_idx]

            # Get user ratings
            user_ratings = train_matrix[user_idx]
            rated_items = np.where(user_ratings > 0)[0]

            if len(rated_items) > 0:
                # Get similarity between this item and items the user has rated
                item_sims = item_similarity[rated_items]

                # If there are similar items
                if np.sum(item_sims > 0) > 0:
                    weighted_sum = np.sum(item_sims * user_ratings[rated_items])
                    weight_sum = np.sum(item_sims)
                    if weight_sum > 0:
                        predictions[i] = weighted_sum / weight_sum
                        covered[i] = True

    # Matrix factorization
    elif method == 'matrix_fact':
        # Replace zeros with NaN to ignore them during factorization
        train_matrix_nan = train_matrix.copy()
        train_matrix_nan[train_matrix_nan == 0] = np.nan

        # Fill NaN with mean rating
        mean_rating = np.nanmean(train_matrix_nan)
        train_matrix_filled = np.nan_to_num(train_matrix_nan, nan=mean_rating)

        # Fit NMF model
        model = NMF(n_components=n_components, init='random', random_state=42, max_iter=200)
        user_factors = model.fit_transform(train_matrix_filled)
        item_factors = model.components_

        # Predict ratings
        reconstructed = np.dot(user_factors, item_factors)

        for i in range(n_test):
            user_idx = test_indices[0][i]
            item_idx = test_indices[1][i]
            actual[i] = test_matrix[user_idx, item_idx]
            predictions[i] = reconstructed[user_idx, item_idx]
            covered[i] = True

    # Calculate error metrics only for covered items
    if np.sum(covered) > 0:
        rmse = np.sqrt(np.mean((predictions[covered] - actual[covered]) ** 2))
        mae = np.mean(np.abs(predictions[covered] - actual[covered]))
        coverage = np.mean(covered) * 100
    else:
        rmse = float('inf')
        mae = float('inf')
        coverage = 0.0

    return rmse, mae, coverage

# Evaluate all models
methods = ['user_cf', 'item_cf', 'matrix_fact', 'baseline']
results = {}

for method in methods:
    rmse, mae, coverage = predict_rating(train_matrix, test_matrix, method=method)
    results[method] = {'RMSE': rmse, 'MAE': mae, 'Coverage (%)': coverage}

# Display results
eval_df = pd.DataFrame(results).T
print("\nModel Evaluation Results:")
print(eval_df)

# Create bar chart for RMSE comparison
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
eval_df['RMSE'].plot(kind='bar', color='skyblue')
plt.title('RMSE by Model')
plt.ylabel('RMSE (lower is better)')
plt.xticks(rotation=45)

# Create bar chart for Coverage comparison
plt.subplot(1, 2, 2)
eval_df['Coverage (%)'].plot(kind='bar', color='lightgreen')
plt.title('Coverage by Model')
plt.ylabel('Coverage (%)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Cell 6: Save Prepared Data and Objects
import os
import joblib

# Create output directories
os.makedirs('models', exist_ok=True)
os.makedirs('data/prepared', exist_ok=True)

print("Saving prepared datasets...")

# Save train and test datasets
train_data.to_csv('data/prepared/train_data.csv', index=False)
test_data.to_csv('data/prepared/test_data.csv', index=False)
print(f"✓ Saved train_data.csv ({train_data.shape[0]} rows)")
print(f"✓ Saved test_data.csv ({test_data.shape[0]} rows)")

# Save the complete filtered dataset
filtered_df.to_csv('data/prepared/filtered_data.csv', index=False)
print(f"✓ Saved filtered_data.csv ({filtered_df.shape[0]} rows)")

print("\nSaving encoders and mappings...")

# Save label encoders
joblib.dump(user_encoder, 'models/user_encoder.pkl')
joblib.dump(beer_encoder, 'models/beer_encoder.pkl')
print("✓ Saved user_encoder.pkl and beer_encoder.pkl")

# Save user and beer mappings for reference
user_mapping = pd.DataFrame({
    'review_profilename': filtered_df['review_profilename'].unique(),
    'user_encoded': user_encoder.transform(filtered_df['review_profilename'].unique())
})
user_mapping.to_csv('models/user_mapping.csv', index=False)

beer_mapping = pd.DataFrame({
    'beer_beerid': filtered_df['beer_beerid'].unique(),
    'beer_encoded': beer_encoder.transform(filtered_df['beer_beerid'].unique())
})
beer_mapping.to_csv('models/beer_mapping.csv', index=False)
print(f"✓ Saved user_mapping.csv ({len(user_mapping)} users)")
print(f"✓ Saved beer_mapping.csv ({len(beer_mapping)} beers)")

# Save feature list for training
feature_columns = [
    'user_encoded', 'beer_encoded', 'review_profilename', 'beer_beerid',
    'review_overall', 'beer_style', 'beer_abv', 'review_aroma', 'review_time'
]
pd.Series(feature_columns).to_csv('models/feature_columns.txt', index=False, header=False)
print("✓ Saved feature_columns.txt")

# Save dataset statistics for reference
stats = {
    'total_users': len(user_encoder.classes_),
    'total_beers': len(beer_encoder.classes_),
    'total_reviews': len(filtered_df),
    'train_size': len(train_data),
    'test_size': len(test_data),
    'unique_styles': filtered_df['beer_style'].nunique(),
    'avg_reviews_per_user': len(filtered_df) / len(user_encoder.classes_),
    'avg_reviews_per_beer': len(filtered_df) / len(beer_encoder.classes_)
}

stats_df = pd.DataFrame(list(stats.items()), columns=['metric', 'value'])
stats_df.to_csv('models/dataset_stats.csv', index=False)
print("✓ Saved dataset_stats.csv")

print("\n" + "="*50)
print("DATA PREPARATION COMPLETE!")
print("="*50)
print(f"📊 Dataset Summary:")
print(f"   • Total Users: {stats['total_users']:,}")
print(f"   • Total Beers: {stats['total_beers']:,}")
print(f"   • Total Reviews: {stats['total_reviews']:,}")
print(f"   • Training Reviews: {stats['train_size']:,}")
print(f"   • Test Reviews: {stats['test_size']:,}")
print(f"   • Unique Beer Styles: {stats['unique_styles']}")

print(f"\n📁 Files saved:")
print("   📂 data/prepared/")
print("      ├── train_data.csv")
print("      ├── test_data.csv")
print("      └── filtered_data.csv")
print("   📂 models/")
print("      ├── user_encoder.pkl")
print("      ├── beer_encoder.pkl")
print("      ├── user_mapping.csv")
print("      ├── beer_mapping.csv")
print("      ├── feature_columns.txt")
print("      └── dataset_stats.csv")

print(f"\n🚀 Ready for model training!")
print("   Next steps:")
print("   1. Create a separate training script")
print("   2. Load the prepared data and encoders")
print("   3. Train your AI recommendation model")
print("   4. Build the Streamlit app")

# Display first few rows of train data to verify
print(f"\n📋 Train Data Sample:")
print(train_data[['review_profilename', 'beer_beerid', 'review_overall',
                  'beer_style', 'beer_abv', 'user_encoded', 'beer_encoded']].head())
```
